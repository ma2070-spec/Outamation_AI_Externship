```python
import numpy as np
import cv2
from PIL import Image
from paddleocr import PaddleOCR

# --- Init (handle old/new param names) ---
try:
    paddle = PaddleOCR(use_textline_orientation=True, lang='en')  # new
except TypeError:
    paddle = PaddleOCR(use_angle_cls=True, lang='en')             # fallback for older installs

# --- Load image ---
image_path = "page_1.png"
img_bgr = cv2.imread(image_path)
if img_bgr is None:
    raise FileNotFoundError(f"Could not read image at {image_path}")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# --- OCR (no cls kwarg) ---
ocr_result = paddle.ocr(img_rgb)  # <-- remove cls=True

# Normalize result to a list of [box, (text, score)]
lines = ocr_result[0] if isinstance(ocr_result, list) and len(ocr_result) else []

# --- Draw boxes ---
canvas = img_bgr.copy()
for line in lines:
    box = line[0]                 # 4 corner points
    text, score = line[1][0], float(line[1][1])

    pts = np.array(box, dtype=np.int32).reshape((-1,1,2))
    cv2.polylines(canvas, [pts], True, (0,255,0), 2)

    x, y = int(box[0][0]), int(box[0][1])
    cv2.putText(canvas, f"{text} ({score:.2f})", (x, y-5),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1, cv2.LINE_AA)

cv2.imwrite("annotated_manual_boxes.png", canvas)
print("Saved annotated_manual_boxes.png")

```
